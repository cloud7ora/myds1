```python
x=np.linspace(0,10,100)

plt.plot(x, np.sin(x))
plt.xlim(-10,22)
plt.ylim(-2,2)
plt.xlabel('THIS IS X)')
plt.xlabel('THIS IS y=six(X)')
plt.title( 'x vs sinx(x)')
```

```python
x=np.linspace(0,20,100)
plt.plot(x, np.sin(x), label='Six(x)')
plt.plot(x, np.cos(x), label='Cox(x)')
plt.xlim(-2,12)
plt.ylim(-2,2)
plt.autoscale()
plt.legend(loc='upper center')
plt.savefig('a1.png', dpi=600)
```

```python
def f(x,y):
  return np.sin(x)** 10+np.sin(10+y*x)*np.cos(x)
x=np.linspace(0,5,50)
y=np.linspace(0,5,50)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
plt.contour(X,Y,Z,colos='Black')
```
